```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

data_path = Path('../data/sample_arctic.csv')
df = pd.read_csv(data_path)
df.head()
```

```python
# Basic info
df.info()
```

```python
# Plot Northern Sea Route transits by year
plt.figure()
df.sort_values('year').plot(x='year', y='nsr_transits', kind='line', legend=False)
plt.title('NSR Transits by Year')
plt.xlabel('Year')
plt.ylabel('Transits')
plt.show()
```

```python
# Plot military upgrades by year
plt.figure()
df.sort_values('year').plot(x='year', y='military_upgrades', kind='bar', legend=False)
plt.title('Military Upgrades (Count) by Year')
plt.xlabel('Year')
plt.ylabel('Upgrades (count)')
plt.show()
```

```python
plt.savefig("../docs/images/arctic_chart.png", dpi=300, bbox_inches="tight")
```
